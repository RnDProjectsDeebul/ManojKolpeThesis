```python
import numpy as np
from scipy.spatial.transform import Rotation as R
```

```python
with open('/home/latai/Downloads/data/test_02/ARCore_sensor_pose.txt') as f:
    lines = f.readlines()
```

```python
quat_values = []
trans_values = []

for i in lines[1:]:
    quat_values.append(list(map(float, i.split()))[1:5])
    trans_values.append(list(map(float, i.split()))[5:])
```

```python
quat_values_np = np.array(quat_values)
trans_values_np = np.array(trans_values)
```

```python
a = np.array([[1, 2, 3],[4, 5, 6],[4, 5, 6]])
b = np.array([[4, 5, 6]])
np.concatenate((a, b.T), axis=1)
```

```python
values = []
for i, j in zip(quat_values_np,trans_values_np) :
    r = R.from_quat(i)
    a = r.as_matrix()
    b = np.array([j])
    c = np.concatenate((a, b.T), axis=1)
    d = np.array([[0.00000000, 0.00000000, 0.00000000, 1.00000000]])
    e = np.concatenate((c, d), axis=0)
    values.append(e.flatten())
    
```
